```python
class Pizza:
#     cheese_toppings = 0
#     pepperoni_toppings = 0
#     mushroom_toppings = 0
#     size = None

        
    def __init__(self,cheese_toppings,pepperoni_toppings,mushroom_toppings,size):
        self.cheese_toppings = cheese_toppings
        self.pepperoni_toppings = pepperoni_toppings
        self.mushroom_toppings = mushroom_toppings
        if size == 1:
            self.size = 'small'
        elif size == 2:
            self.size = 'medium'
        elif size == 3:
            self.size = 'large'
        else:
            size = None
            
        
    def get_cheese_toppings(self):
        return self.cheese_toppings
    
    def set_cheese_toppings(self,cheese_toppings):
        self.cheese_toppings = cheese_toppings
    
    def get_pepperoni_toppings(self):
        return self.pepperoni_toppings
    
    def set_pepperoni_toppings(self,pepperoni_toppings):
        self.pepperoni_toppings = pepperoni_toppings
    
    def get_mushroom_toppings(self):
        return self.mushroom_toppings
    
    def set_mushroom_toppings(self,mushroom_toppings):
        self.mushroom_toppings = mushroom_toppings    
        
    def calcCost(self):
        if self.size == 'small':
            return 10+(self.cheese_toppings*2)+(self.pepperoni_toppings*2)+(self.mushroom_toppings*2)
        elif self.size == 'medium':
            return 12+(self.cheese_toppings*2)+(self.pepperoni_toppings*2)+(self.mushroom_toppings*2)
        elif self.size == 'large':
            return 14+(self.cheese_toppings*2)+(self.pepperoni_toppings*2)+(self.mushroom_toppings*2)
        else:
            return (self.cheese_toppings*2)+(self.pepperoni_toppings*2)+(self.mushroom_toppings*2)
        
    def __str__(self):
        return f"""Size of pizza : {self.size} 
                   Cheese Toppings : {str(self.cheese_toppings)}
                   Pepperoni Toppings : {str(self.pepperoni_toppings)}
                   Mushroom Toppings : {str(self.mushroom_toppings)}"""
```

```python
class DeluxePizza:

    numberOfPizzas = 0
        
    def __init__(self,cheese_toppings,pepperoni_toppings,mushroom_toppings,size,veggieTopping,stuffedWithCheese):
        self.cheese_toppings = cheese_toppings
        self.pepperoni_toppings = pepperoni_toppings
        self.mushroom_toppings = mushroom_toppings
        self.veggieTopping = veggieTopping
        if stuffedWithCheese == 1:
            self.stuffedWithCheese = True
        elif stuffedWithCheese == 2:
            self.stuffedWithCheese = False
        else:
            self.stuffedWithCheese = False
            
        if size == 1:
            self.size = 'small'
        elif size == 2:
            self.size = 'medium'
        elif size == 3:
            self.size = 'large'
        else:
            size = None
        
#         DeluxePizza.numberOfPizzas += 1
            
    def get_size(self):
        return self.size
    
    def set_size(self,size):
        if size == 1:
            self.size = 'small'
        elif size == 2:
            self.size = 'medium'
        elif size == 3:
            self.size = 'large'
        else:
            size = None
    
    def get_cheese_toppings(self):
        return self.cheese_toppings
    
    def set_cheese_toppings(self,cheese_toppings):
        self.cheese_toppings = cheese_toppings
    
    def get_pepperoni_toppings(self):
        return self.pepperoni_toppings
    
    def set_pepperoni_toppings(self,pepperoni_toppings):
        self.pepperoni_toppings = pepperoni_toppings
    
    def get_mushroom_toppings(self):
        return self.mushroom_toppings
    
    def set_mushroom_toppings(self,mushroom_toppings):
        self.mushroom_toppings = mushroom_toppings    
        
    def get_stuffedWithCheese(self):
        return self.stuffedWithCheese
    
    def set_stuffedWithCheese(self,stuffedWithCheese):
        if stuffedWithCheese == 1:
            self.stuffedWithCheese = True
        elif stuffedWithCheese == 2:
            self.stuffedWithCheese = False
        else:
            self.stuffedWithCheese = False
        
    def get_veggieTopping(self):
        return self.veggieTopping
    
    def set_veggieTopping(self,veggieTopping):
        self.veggieTopping = veggieTopping    
        
    def numberOfPizzas(self):
        return self.numberOfPizzas
        
    def calcCost(self):
        if self.size == 'small':
            
            if self.stuffedWithCheese == True:
                return 10+2+(self.cheese_toppings*2)+(self.pepperoni_toppings*2)+(self.mushroom_toppings*2)+(self.veggieTopping*3)
            else:
                return 10+(self.cheese_toppings*2)+(self.pepperoni_toppings*2)+(self.mushroom_toppings*2)+(self.veggieTopping*3)
        
        elif self.size == 'medium':
            
            if self.stuffedWithCheese == True:
                return 12+4+(self.cheese_toppings*2)+(self.pepperoni_toppings*2)+(self.mushroom_toppings*2)+(self.veggieTopping*3)
            else:
                return 12+(self.cheese_toppings*2)+(self.pepperoni_toppings*2)+(self.mushroom_toppings*2)+(self.veggieTopping*3)
                
        elif self.size == 'large':
            
            if self.stuffedWithCheese == True:
                return 14+6+(self.cheese_toppings*2)+(self.pepperoni_toppings*2)+(self.mushroom_toppings*2)+(self.veggieTopping*3)
            else:
                return 14+(self.cheese_toppings*2)+(self.pepperoni_toppings*2)+(self.mushroom_toppings*2)+(self.veggieTopping*3)
        
        else:
            return (self.cheese_toppings*2)+(self.pepperoni_toppings*2)+(self.mushroom_toppings*2)+(self.veggieTopping*3)
        
    def __str__(self):
        return f"""Size of pizza : {self.size} 
                   Cheese filled dough: {self.stuffedWithCheese}
                   Cheese Toppings : {str(self.cheese_toppings)}
                   Pepperoni Toppings : {str(self.pepperoni_toppings)}
                   Mushroom Toppings : {str(self.mushroom_toppings)}
                   Veggie Topping : {str(self.veggieTopping)}"""
```

```python
todaysPizzas = []

def pizza_order():
    pswd = "deluxepizza"
    
    print("\n")
    count = 0
    while count<3:
        pd = input("Enter password to enter pizza details: ")
        if pd == pswd:
            number_of_pizzas = int(input("How many pizzas you want to enter ? : "))
            print("\n")
            if number_of_pizzas <=10:
                for pizza in range(number_of_pizzas):
                    print(f"Enter the order details of pizza {pizza+1} : ")
                    size = int(input("Enter the size of pizza - 1. Small 2. Meduim 3. Large :"))
                    stuffedWithCheese = int(input("Cheese filled dough - 1. Yes 2. No"))
                    cheese_toppings = int(input("Number of cheese Topping : "))
                    pepperoni_toppings = int(input("Number of pepproni toppings : "))
                    mushroom_toppings = int(input("Number of mushroom toppings : "))
                    veggieTopping = int(input("Number of veggie toppings : "))
                    print("\n")
                    todaysPizzas.append(DeluxePizza(cheese_toppings,pepperoni_toppings,mushroom_toppings,size,veggieTopping,stuffedWithCheese))
                break        
            else:
                print("pizza limit crossed")
        else:
            print("Wrong password\n")
            count+=1

            
```

```python
def pizza_update():
    pswd = "deluxepizza"
    
    print("\n")
    count = 0
    while count<3:
        pd = input("Enter password to Update pizza details: ")
        print("\n")
        if pd == pswd:
            count1 = 0
            for pizza in todaysPizzas:
                count1+=1
                print(f"{count1}. : "+pizza.__str__())
                print("\n")
            number_of_pizza = int(input("Which pizza number you want to update ? : "))
            print("\n")  
            print(f"Enter the order details of pizza {number_of_pizza} to update : ")
            pizza = todaysPizzas[number_of_pizza-1]
            size = int(input("Enter the size of pizza - 1. Small 2. Meduim 3. Large :"))
            stuffedWithCheese = int(input("Cheese filled dough - 1. Yes 2. No"))
            cheese_toppings = int(input("Number of cheese Topping : "))
            pepperoni_toppings = int(input("Number of pepproni toppings : "))
            mushroom_toppings = int(input("Number of mushroom toppings : "))
            veggieTopping = int(input("Number of veggie toppings : "))
            print("\n")
            
            pizza.set_size(size)
            pizza.set_cheese_toppings(cheese_toppings)
            pizza.set_pepperoni_toppings(pepperoni_toppings)
            pizza.set_mushroom_toppings(mushroom_toppings)
            pizza.set_veggieTopping(veggieTopping)
            pizza.set_stuffedWithCheese(stuffedWithCheese)
            print("\n")
            print("Pizza Details Updated !! ")
            print(pizza.__str__())
            print("\n")
            break        
            
        else:
            print("Wrong password\n")
            count+=1
```

```python
def pizzasOfSize():
    size = int(input("Enter which size of pizzas you want to see : 1. Small 2. Medium 3. Large 4. Back to main menu : "))
    if size == 1:
        for pizza in todaysPizzas:
            if pizza.get_size()=='small':
                print(pizza.__str__())
    elif size == 2:
        for pizza in todaysPizzas:
            if pizza.get_size()=='medium':
                print(pizza.__str__())
    elif size == 3:
        for pizza in todaysPizzas:
            if pizza.get_size()=='large':
                print(pizza.__str__())
    elif size == 4:
        main()
    
    else:
        print("Wrong size entered !! ")
```
